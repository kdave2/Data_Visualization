```python

"""
Demo using fontdict to control style of text and labels.
"""

import matplotlib.pyplot as plt

#define font
font = {'family' : 'serif',
        'color'  : 'darkred',
        'weight' : 'normal',
        'size'   : 16,
        }


#set the rane of the x_axis to be from 0 to 12
canvas = plt.gca()
canvas.set_xlim([0,12])

# some random data
x = [2,4,6,8,10]
y = [10,45,60,80,130]

plt.plot(x, y)

#labels
labels=['A','B','C','D','E']

#use the labels as the marks on the x-axis, rotate them by 45 degrees.
plt.xticks(x, labels, rotation=45)

#set the title and labels of the plot, using the predfined font style
plt.title('Fonts Example', fontdict=font)

plt.xlabel('X Label', fontdict=font)
plt.ylabel('Y Label', fontdict=font)

plt.show()
```

```python


"""
Simple demo of a horizontal bar chart.
"""
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt


# The labels of the bars
labels = ('Tom', 'JOhn', 'Harry', 'Slim', 'Jim')

#The positions of each bar on the y-axis (5 bars in this case)
y_pos=[0 ,1 ,2, 3, 4]

#the age that corresponds to each user
ages =[20,20,32,56,89] 

#Make random error bars. np.random.rand(x) returns a table of x numbers between 0 and 1
error = np.random.rand(len(labels))

#barh makes the histogram. Alpha tunes the darkness of the color
plt.barh(y_pos, ages, xerr=error, align='center', color='red', alpha=0.4) 

#add the labels
plt.yticks(y_pos, labels)

#Add the label of the x axis
plt.xlabel('Age')

#Add the title of the plot
plt.title('Age Histogram')
plt.show()
#save the plot
plt.savefig('horizontal.png', format='png')
```

```python
# -*- coding: utf-8 -*-

"""
Demo of a basic pie chart plus a few additional features.
"""

import matplotlib.pyplot as plt


# The slices will be ordered and plotted counter-clockwise.
labels = 'France', 'England', 'Germany', 'USA'

#size of each slice. These don;t have to sum up to 100. Python normalizes values automatically.
sizes = [100 ,100, 405, 120]

#color of each slice
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

#The explode value determines if a slice should be sticking out and to what extent.
explode = (0, 0.5, 0, 0)

#make the chart. Autopct is used to also add the percentage on top of the slice.
#You can add/remove the shadow with the shadow feature.
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%.1f%%', shadow=True)
        
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')

plt.show()
```

```python

"""
Simple demo with multiple subplots.
"""
import numpy as np
import matplotlib.pyplot as plt

#random data for two different plots
x1=[1,2,3,4,5]
y1 =[10,20,30,40,60]

x2=[1,2,3,4,5]
y2=[20,40,10,50,60]

#a = plt.gca()
#a.set_ylim([0,10])

#subplot accepts two parameters: number of rows (how many plots in a row), 
#number of colums(how many plots in a column) and the plot ID (determines placement on the figure)
plt.subplot(2, 1, 1) 
plt.plot(x1, y1, 'yo-')#yo- means use yellow circles and a solid line
plt.xlabel('Plot 1')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.--')# r.-- means use red dots and a dashed line
plt.xlabel('Plot 2')

"""
you can find all possible line styles here:
http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
"""

plt.show()
```

```python
# -*- coding: utf-8 -*-


"""
Simple demo of a vertical bar chart.
"""
import matplotlib.pyplot as plt


#set the rane of the x_axis to be from 0 to 10
canvas = plt.gca()
canvas.set_ylim([0,10])

#number of bins to use
num_bins = 5

#some random data
data=[10,20,3,4,5,6,3,3,65,6,3,43,43,34]

#hist makes the histogram
#alpha tunes the darkness of the bars
n, bins, patches = plt.hist(data, num_bins, rwidth=0.8,facecolor='blue', alpha=0.9)


plt.xticks(bins)

#add labels and title
plt.xlabel('Frequency')
plt.ylabel('Bins')
plt.title('My Histogram')

# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)

plt.show()
```
